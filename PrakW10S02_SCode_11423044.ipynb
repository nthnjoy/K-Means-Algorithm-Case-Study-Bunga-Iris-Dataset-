```python
# Nathanael T.J. Tampubolon
# 11423044

import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

X = iris.select_dtypes(include=['float64', 'int64'])

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, inertia, 'bo-')
plt.title('Metode Elbow untuk Menentukan Jumlah Cluster Optimal')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
#Nathanael T.J. Tampubolon
#11423044
kmeans = KMeans(n_clusters=3, random_state=42)
iris['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Nathanael T.J. Tampubolon
# 11423044

from sklearn.metrics import silhouette_score  

score = silhouette_score(X_scaled, iris['Cluster'])
print(f"Silhouette Score: {score:.3f}")
```

```python
#Nathanael T.J. Tampubolon
#11423044
plt.figure(figsize=(8,5))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1],
                hue=iris['Cluster'], palette='Set2')
plt.title('Visualisasi Clustering (2D)')
plt.xlabel('SepalLengthCm (scaled)')
plt.ylabel('SepalWidthCm (scaled)')
plt.legend(title='Cluster')
plt.show()
```

```python
#Nathanael T.J. Tampubolon
#11423044
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(X_scaled[:, 0], X_scaled[:, 1], X_scaled[:, 2],
                     c=iris['Cluster'], cmap='Set2', s=60)

ax.set_title('Visualisasi Clustering (3D)')
ax.set_xlabel('SepalLengthCm (scaled)')
ax.set_ylabel('SepalWidthCm (scaled)')
ax.set_zlabel('PetalLengthCm (scaled)')
plt.legend(*scatter.legend_elements(), title="Cluster")
plt.show()
```

```python
# Nathanael T.J. Tampubolon
# 11423044

import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


iris_data = load_iris()
iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
iris['species'] = iris_data.target  #  tambahkan kolom species (label asli)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(iris_data.data)


kmeans = KMeans(n_clusters=3, random_state=42)
iris['Cluster'] = kmeans.fit_predict(X_scaled)


print("Data Hasil Clustering:")
print(iris.head())

print("\nPerbandingan Cluster vs Label Asli:")
print(pd.crosstab(iris['Cluster'], iris['species']))
```

```python
#Nathanael T.J. Tampubolon
#11423044
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.decomposition import PCA

iris = pd.read_csv("Iris.csv")

X = iris.iloc[:, [0,1,2,3]].values
feature_names = iris.columns[0:4].tolist()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k = 4
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
labels = kmeans.fit_predict(X_scaled)
sil_avg = silhouette_score(X_scaled, labels)
sil_samples = silhouette_samples(X_scaled, labels)

print(f"Silhouette Score untuk k={k}: {sil_avg:.4f}")
```

```python
#Nathanael T.J. Tampubolon
#11423044
#Visualisasi 2D

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
plt.figure(figsize=(7,6))
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=labels, s=50)
plt.title(f"PCA 2D projection (k={k})")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(*scatter.legend_elements(), title="Cluster")
plt.show()
```

```python
#Nathanael T.J. Tampubolon
#11423044
#Visualisasi 3D
fig = plt.figure(figsize=(8,6))
ax3 = fig.add_subplot(111, projection='3d')
scatter3 = ax3.scatter(X_scaled[:,0], X_scaled[:,1], X_scaled[:,2], c=labels, s=50)
ax3.set_title(f"3D Scatter (first 3 scaled features) - k={k}")
ax3.set_xlabel(feature_names[0])
ax3.set_ylabel(feature_names[1])
ax3.set_zlabel(feature_names[2])
plt.legend(*scatter3.legend_elements(), title="Cluster")
plt.show()
```
